```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("time_series.csv")

df["sale_month"] = pd.to_datetime(df["sale_month"])
df = df.sort_values("sale_month")

plt.figure()

plt.plot(df["sale_month"], df["total_profit"], label="Profit")
plt.plot(df["sale_month"], df["total_units"], label="Units Sold")

plt.xlabel("Month")
plt.ylabel("Value")
plt.title("Profit and Units Over Time")

# Shows every 3rd month
plt.xticks(df["sale_month"][::3], rotation=45)  
plt.legend()
plt.tight_layout()

plt.show()
```
